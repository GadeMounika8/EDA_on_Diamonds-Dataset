# Data Cleaning:

```python
# gives number of rows and columns
df.shape
```

```python
# Top 10 Records
df.head(10)
```

```python
# number of duplicate values in a DataFrame
df.duplicated().sum()
```

```python
# dropping the duplicates from the DataFrame
df.drop_duplicates(inplace = True)
```

```python
df.isnull().sum()
```

```python
# removing outliers in carat column by using IQR Method
Q1 = df['carat'].quantile(0.25)
Q3 = df['carat'].quantile(0.75)
IQR = Q3 - Q1
bmi_lower_bound = Q1 - 1.5*IQR
bmi_upper_bound = Q3 + 1.5*IQR
df = df[(df['carat'] > bmi_lower_bound)&(df['carat']<bmi_upper_bound)]
print(df.shape)
```

```python
# removing outliers in depth column by using IQR Method
Q1 = df['depth'].quantile(0.25)
Q3 = df['depth'].quantile(0.75)
IQR = Q3 - Q1
bmi_lower_bound = Q1 - 1.5*IQR
bmi_upper_bound = Q3 + 1.5*IQR
df = df[(df['depth'] > bmi_lower_bound)&(df['depth']<bmi_upper_bound)]
print(df.shape)
```

```python
# removing outliers in table column by using IQR Method
Q1 = df['table'].quantile(0.25)
Q3 = df['table'].quantile(0.75)
IQR = Q3 - Q1
bmi_lower_bound = Q1 - 1.5*IQR
bmi_upper_bound = Q3 + 1.5*IQR
df = df[(df['table'] > bmi_lower_bound)&(df['table']<bmi_upper_bound)]
print(df.shape)
```

```python
# removing outliers in x column by using IQR Method
Q1 = df['x'].quantile(0.25)
Q3 = df['x'].quantile(0.75)
IQR = Q3 - Q1
bmi_lower_bound = Q1 - 1.5*IQR
bmi_upper_bound = Q3 + 1.5*IQR
df = df[(df['x'] > bmi_lower_bound)&(df['x']<bmi_upper_bound)]
print(df.shape)
```

```python
# removing outliers in y column by using IQR Method
Q1 = df['y'].quantile(0.25)
Q3 = df['y'].quantile(0.75)
IQR = Q3 - Q1
bmi_lower_bound = Q1 - 1.5*IQR
bmi_upper_bound = Q3 + 1.5*IQR
df = df[(df['y'] > bmi_lower_bound)&(df['y']<bmi_upper_bound)]
print(df.shape)
```

```python
# removing outliers in z column by using IQR Method
Q1 = df['z'].quantile(0.25)
Q3 = df['z'].quantile(0.75)
IQR = Q3 - Q1
bmi_lower_bound = Q1 - 1.5*IQR
bmi_upper_bound = Q3 + 1.5*IQR
df = df[(df['z'] > bmi_lower_bound)&(df['z']<bmi_upper_bound)]
print(df.shape)
```
